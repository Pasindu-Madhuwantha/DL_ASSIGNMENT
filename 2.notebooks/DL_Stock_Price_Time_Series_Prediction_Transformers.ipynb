#Data Extraction

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Load the dataset from the CSV file
df = pd.read_csv('/content/drive/MyDrive/DATA.csv')  # Replace 'your_dataset.csv' with the actual file name

# Rename the columns to make them more accessible
df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'NASDAQ', 'NYSE', 'S&P 500', 'FTSE100',
              'NIKKI225', 'BSE SENSEX', 'RUSSELL2000', 'HENG SENG', 'SSE', 'Crude Oil', 'Gold', 'VIX',
              'USD index', 'Amazon', 'Google', 'Microsoft']

# Convert the 'Date' column to a datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Choose a stock symbol (Gold in this case)
target = 'Gold'

# Choose a date range
train_start_date = datetime(2010, 7, 1).date()
train_end_date = datetime(2018, 6, 18).date()
test_start_date = train_end_date

# Plot the Gold stock price
plt.figure(figsize=(16, 4))
plt.plot(df['Date'], df[target], label=f'{target} stock price', linewidth=2)
plt.axvline(x=train_end_date, color='r', linestyle='--', label=f'Training-Test Split Date ({train_end_date})')
plt.legend()
plt.title(f'{target} stock price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
